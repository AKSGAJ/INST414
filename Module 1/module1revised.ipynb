```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

df = pd.read_csv("Most Streamed Spotify Songs 2024.csv", encoding='latin1')

columns = ['Track', 'Spotify Streams', 'TikTok Posts', 'TikTok Views', 'Release Date']
df_selected = df[columns].dropna()

for col in ['Spotify Streams', 'TikTok Posts', 'TikTok Views']:
    df_selected[col] = df_selected[col].replace(',', '', regex=True).astype(int)


```

```python
# Use only the first 20 songs to keep the x-axis readable
df_plot = df_selected.head(20)

plt.figure(figsize=(14, 6))
plt.plot(df_plot['Track'], df_plot['Spotify Streams'], label='Spotify Streams', color='green', marker='o')
plt.plot(df_plot['Track'], df_plot['TikTok Posts'], label='TikTok Posts', color='blue', marker='o')
plt.plot(df_plot['Track'], df_plot['TikTok Views'], label='TikTok Views', color='purple', marker='o')

plt.title('Spotify Streams vs TikTok Engagement (Top 20 Songs)')
plt.xlabel('Track')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

print(df_selected[['Track', 'Release Date']].head(20))
```

```python
df_sorted = df_selected.sort_values(by='Spotify Streams', ascending=False)
df_plot = df_sorted.head(20)

plt.figure(figsize=(14, 6))
plt.plot(df_plot['Track'], df_plot['Spotify Streams'], label='Spotify Streams', color='green', marker='o')
plt.plot(df_plot['Track'], df_plot['TikTok Posts'], label='TikTok Posts', color='blue', marker='o')
plt.plot(df_plot['Track'], df_plot['TikTok Views'], label='TikTok Views', color='purple', marker='o')

plt.title('Spotify Streams vs TikTok Engagement (Top 20 Songs)')
plt.xlabel('Track')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

print(df_sorted[['Track', 'Release Date']].head(20))
```
